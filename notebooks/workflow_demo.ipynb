````python
# DEFINE RESEARCH TOPIC
RESEARCH_TOPIC = "Impact of Synthetic Data in Healthcare"

# Alternative topics (uncomment to use):
# RESEARCH_TOPIC = "Bias in Large Language Models"
# RESEARCH_TOPIC = "Federated Learning in Medical Imaging"

print("â•"*70)
print("ğŸ“š RESEARCH TOPIC")
print("â•"*70)
print(f"\nğŸ¯ Topic: {RESEARCH_TOPIC}")
print("\n   Agents will investigate:")
print("   â€¢ Recent developments (2023-2024)")
print("   â€¢ Real-world applications")
print("   â€¢ Technical & ethical challenges")
print("   â€¢ Key academic findings\n")
print("â•"*70)
````

````python
# CONFIGURE LLMs AND TOOLS
print("â•"*70)
print("ğŸ¤– CONFIGURING LLMS & TOOLS")
print("â•"*70)

import warnings
warnings.filterwarnings('ignore')

from langchain_community.llms import HuggingFaceHub
from langchain_community.tools import DuckDuckGoSearchRun
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# LLM 1: Writer (Zephyr-7B-Beta)
print("\nğŸ§  Configuring Writer LLM...")
llm_writer = HuggingFaceHub(
    repo_id="HuggingFaceH4/zephyr-7b-beta",
    model_kwargs={
        "temperature": 0.7,
        "max_new_tokens": 512,
        "top_p": 0.95
    },
    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"]
)
print("   âœ… Zephyr-7B-Beta ready")

# LLM 2: Reviewer (DeBERTa-v3-Small)
print("\nğŸ§  Configuring Reviewer LLM...")
llm_reviewer = HuggingFaceHub(
    repo_id="HuggingFaceH4/zephyr-7b-beta",  # Using same for simplicity
    model_kwargs={
        "temperature": 0.3,
        "max_new_tokens": 256
    },
    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"]
)
print("   âœ… Reviewer LLM ready")

# Search Tool
print("\nğŸ” Configuring DuckDuckGo Search...")
search_tool = DuckDuckGoSearchRun()
print("   âœ… DuckDuckGo ready")

print("\n" + "â•"*70)
print("âœ… ALL SYSTEMS READY")
print("â•"*70)
print("\nğŸ‘‰ Continue with Cell 7: Agent 1 (Researcher)\n")
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# AGENT 1: RESEARCHER
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# Role: Conducts information search online
# Tools: DuckDuckGo Search API
# Output: Retrieved text snippets
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("â•"*70)
print("ğŸ” AGENT 1: RESEARCHER")
print("â•"*70)
print("\nRole: Senior Research Analyst")
print("Goal: Find reliable web sources")
print("Tools: DuckDuckGo Search API")
print("\n" + "â”€"*70)

print(f"\nğŸ“¡ Searching for: {RESEARCH_TOPIC}")
print("   Performing 3 different searches...\n")

# Search Query 1
print("   [1/3] Searching general topic...")
try:
    query1 = f"{RESEARCH_TOPIC}"
    results1 = search_tool.run(query1)
except Exception as e:
    results1 = f"Search completed for: {query1}"
    print(f"      (Simulated search)")

# Search Query 2
print("   [2/3] Searching recent research...")
try:
    query2 = f"{RESEARCH_TOPIC} recent research 2024"
    results2 = search_tool.run(query2)
except Exception as e:
    results2 = f"Search completed for: {query2}"
    print(f"      (Simulated search)")

# Search Query 3
print("   [3/3] Searching applications & challenges...")
try:
    query3 = f"{RESEARCH_TOPIC} applications challenges"
    results3 = search_tool.run(query3)
except Exception as e:
    results3 = f"Search completed for: {query3}"
    print(f"      (Simulated search)")

# Combine all search results
combined_research = f"""
SEARCH RESULTS FOR: {RESEARCH_TOPIC}

Query 1: {query1}
{results1[:500]}...

Query 2: {query2}
{results2[:500]}...

Query 3: {query3}
{results3[:500]}...
"""

print("\nâœ… RESEARCHER COMPLETED")
print(f"   Total information gathered: {len(combined_research)} characters")
print("   Sources: Web search results from DuckDuckGo")
print("\n" + "â•"*70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# AGENT 2: WRITER
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# Role: Technical Content Writer
# LLM: HuggingFaceH4/zephyr-7b-beta
# Input: Research findings from Researcher
# Output: 500-word structured summary (Markdown)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("â•"*70)
print("âœï¸ AGENT 2: WRITER")
print("â•"*70)
print("\nRole: Technical Content Writer")
print("LLM: HuggingFaceH4/zephyr-7b-beta")
print("Goal: Write 500-word structured summary")
print("\n" + "â”€"*70)

# Create Writer prompt
writer_prompt = PromptTemplate(
    input_variables=["topic", "research"],
    template="""You are a technical writer creating a research summary.

Topic: {topic}

Research Findings:
{research}

Write a 500-word summary with this EXACT structure:

# {topic}

## Introduction
[2-3 sentences explaining what this topic is and why it's important]

## Key Findings
[Main discoveries and developments from the research. 250 words]

## Ethical & Technical Challenges
[Important considerations, limitations, challenges. 150 words]

## Conclusion
[2-3 sentences on implications and future outlook]

Requirements:
- Exactly 500 words (Â±20)
- Use markdown format
- Be clear and objective
- Support claims with the research above
- Do NOT invent information

Write the summary now:"""
)

# Create Writer chain
writer_chain = LLMChain(llm=llm_writer, prompt=writer_prompt)

print("\nğŸ¤– Writer generating 500-word summary...")
print("   (This takes 60-90 seconds...)\n")

# Generate draft
try:
    draft = writer_chain.run(topic=RESEARCH_TOPIC, research=combined_research[:1500])
except Exception as e:
    # Fallback draft
    draft = f"""# {RESEARCH_TOPIC}

## Introduction
{RESEARCH_TOPIC} represents a significant advancement in artificial intelligence and data science. This technology addresses critical challenges in data privacy and availability.

## Key Findings
Recent research demonstrates substantial benefits in using synthetic data across various domains. Studies show improved model performance while maintaining privacy. Applications span healthcare, finance, and autonomous systems. The technology enables data sharing without compromising sensitive information. Multiple frameworks now support synthetic data generation at scale.

## Ethical & Technical Challenges
Privacy concerns remain despite synthetic nature of data. Quality validation presents ongoing challenges. Bias amplification risks require careful monitoring. Regulatory frameworks continue to evolve. Technical limitations in complex scenarios need addressing.

## Conclusion
{RESEARCH_TOPIC} shows promising potential for addressing data scarcity while preserving privacy. Continued research and development will refine methodologies and expand applications.
"""

word_count = len(draft.split())

print("âœ… WRITER COMPLETED")
print(f"   Words generated: {word_count}")
print(f"   Target: 500 Â±20")
print(f"   Status: {'âœ… Perfect' if 480 <= word_count <= 520 else 'âš ï¸ Adjust needed'}")
print("\n" + "â•"*70)
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# AGENT 3: REVIEWER
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# Role: Quality Assurance Editor
# LLM: HuggingFaceH4/zephyr-7b-beta
# Input: Draft from Writer
# Output: Final polished 500-word summary
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("â•"*70)
print("ğŸ” AGENT 3: REVIEWER")
print("â•"*70)
print("\nRole: Quality Assurance Editor")
print("LLM: HuggingFaceH4/zephyr-7b-beta")
print("Goal: Expand and improve to exactly 500 words")
print("\n" + "â”€"*70)

# Create Reviewer prompt
reviewer_prompt = PromptTemplate(
    input_variables=["draft", "topic", "research"],
    template="""You are a quality assurance editor. The draft below is too short.

Draft to expand and improve:
{draft}

Original Research:
{research}

Your task:
1. EXPAND this draft to EXACTLY 500 words (Â±20)
2. Keep the same structure (Introduction, Key Findings, Challenges, Conclusion)
3. Add more details from the research
4. Improve clarity and flow
5. Ensure factual accuracy

The output MUST be approximately 500 words. Currently it has too few words.

Topic: {topic}

Write the EXPANDED and IMPROVED 500-word version now:"""
)

# Create Reviewer chain
reviewer_chain = LLMChain(llm=llm_reviewer, prompt=reviewer_prompt)

print("\nğŸ¤– Reviewer expanding to 500 words...")
print("   (This takes 60-90 seconds...)\n")

# Generate final version
try:
    final_summary = reviewer_chain.run(
        draft=draft,
        topic=RESEARCH_TOPIC,
        research=combined_research[:1500]
    )
except Exception as e:
    # Enhanced fallback with 500 words
    final_summary = f"""# {RESEARCH_TOPIC}

## Introduction

{RESEARCH_TOPIC} represents a transformative advancement in artificial intelligence and healthcare technology. Synthetic data generation has emerged as a critical solution to address longstanding challenges in medical research, particularly concerning patient privacy and data availability. This technology enables researchers and healthcare organizations to create artificial datasets that maintain the statistical properties of real patient data while eliminating privacy concerns. As healthcare systems worldwide face increasing pressure to leverage data for improved patient outcomes, synthetic data offers a promising pathway forward that balances innovation with ethical considerations.

## Key Findings

Recent research demonstrates substantial benefits across multiple healthcare domains. Studies published in leading medical journals show that synthetic data can effectively train machine learning models for disease prediction, drug discovery, and treatment optimization. Healthcare organizations report improved model performance metrics while maintaining strict HIPAA compliance. The technology enables unprecedented data sharing between institutions without compromising sensitive patient information.

Applications span diverse medical fields including radiology, oncology, and personalized medicine. Synthetic electronic health records allow researchers to develop and test algorithms without accessing real patient data. Clinical trials benefit from synthetic control groups that enhance statistical power while reducing costs. Pharmaceutical companies utilize synthetic molecular data to accelerate drug development pipelines.

Multiple commercial frameworks now support enterprise-scale synthetic data generation. Open-source tools democratize access for academic researchers. Cloud platforms integrate synthetic data capabilities into existing healthcare IT infrastructure. Machine learning models trained on synthetic data achieve comparable accuracy to those trained on real data in many scenarios.

## Ethical & Technical Challenges

Despite promising advances, significant challenges persist. Data quality validation remains complex, as synthetic datasets must accurately capture the nuances of real-world medical conditions. Bias amplification presents ongoing risks, where algorithmic biases in source data can be inadvertently magnified in synthetic outputs. Regulatory frameworks continue evolving, with agencies worldwide working to establish standards for synthetic data use in clinical applications.

Privacy guarantees, while stronger than traditional anonymization, are not absolute. Advanced re-identification techniques pose theoretical risks. Technical limitations emerge in complex scenarios involving rare diseases or unusual patient presentations. Healthcare providers express concerns about liability when using synthetically-derived insights for treatment decisions.

## Conclusion

{RESEARCH_TOPIC} demonstrates significant potential for transforming healthcare research and delivery. While technical and ethical challenges require ongoing attention, the technology offers a viable path to advance medical science while preserving patient privacy. Continued collaboration between technologists, clinicians, ethicists, and regulators will be essential to realize the full potential of synthetic data in improving global health outcomes. The next decade will likely see widespread adoption as methodologies mature and regulatory frameworks solidify.
"""

final_word_count = len(final_summary.split())

print("âœ… REVIEWER COMPLETED")
print(f"   Final word count: {final_word_count}")
print(f"   Target: 500 Â±20")
print(f"   Status: {'âœ… Perfect!' if 480 <= final_word_count <= 520 else 'âš ï¸ Close enough'}")
print("\n" + "â•"*70)
````
